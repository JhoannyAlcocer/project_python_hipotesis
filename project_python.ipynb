## Prueba de hipotesis

Hipotesis: "La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos"

Dado que los datos provistos no tienen registros de las condiciones climaticas los dias sabados, se propone una nueva hipotesis: 

"¿La duración promedio de los viajes cambia dependiendo de si las condiciones climáticas son “Good” o “Bad”?"

Hipotesis nula (H₀):
La duración promedio de los viajes es igual bajo condiciones “Good” y “Bad”.

Hipótesis alternativa (H₁):
La duración promedio de los viajes es diferente entre condiciones “Good” y “Bad”.

#### Carga y estudio de datos 

```python
import pandas as pd
import numpy as np
from scipy import stats as st
```

```python
df= pd.read_csv('/datasets/project_sql_result_07.csv')
```

```python
print(df.head())
df.info()
```

#### Cambio de tipos de datos a datetime 

```python
df['start_ts'] = pd.to_datetime(df['start_ts'])
```

```python
df.info()
```

#### Separamos los datos de acuerdo a la condicion climatica 'Good' o 'Bad'

```python
good_weather = df[df['weather_conditions'] == 'Good']['duration_seconds']
bad_weather = df[df['weather_conditions'] == 'Bad']['duration_seconds']
```

#### Usamos el T-test de dos muestras independientes.

```python
alpha= 0.05
```

```python
result = st.ttest_ind(good_weather, bad_weather, equal_var=False)
```

```python
print("p-valor:", result.pvalue)
```

```python
if result.pvalue < alpha:
    print("Rechazamos H₀: la duración promedio de los viajes cambia según el clima.")
else:
    print("No se puede rechazar H₀: no hay evidencia de una diferencia significativa.")
```

### Conclusion

Para comprobrar la hipotesis de si la duración promedio de los viajes cambia dependiendo las condiciones climaticas, y con los datos provistos, se utilizo un nivel de significancia de 0.05, esto significa que estamos dispuestos a aceptar un 5% de probabilidad de rechazar la hipótesis nula cuando en realidad es verdadera (error tipo I).

Usamos una prueba t de dos muestras independientes, ya que comparamos las medias de dos grupos independientes (Good y Bad) y debido a que las varianzas pueden ser distintas, así que usamos equal_var=False.

Con los calculos realizados anteriormente, concluimos que la duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O’Hare cambia cuando llueve.
